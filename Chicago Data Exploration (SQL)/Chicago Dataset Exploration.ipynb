## Problems to solve

### Problem 1

##### How many community areas in Chicago have a hardship index greater than 50.0?

### Problem 2

##### What is the maximum value of hardship index in this dataset?

### Problem 3

##### Which community area has the highest hardship index?

```python
%%sql 

select community_area_name 
from CENSUS_DATA 
where hardship_index = (select max(hardship_index) from CENSUS_DATA)
```

### Problem 4

##### Which Chicago community areas have per-capita incomes greater than $60,000?

```python
%%sql 

SELECT community_area_name 
FROM CENSUS_DATA 
WHERE per_capita_income > 60000
```

### Problem 5

##### Create a scatter plot using the variables `per_capita_income` and `hardship_index` and explain the correlation between the two variables.


```python
income_vs_hardship = %sql SELECT per_capita_income, hardship_index FROM CENSUS_DATA;
```

```python
data = income_vs_hardship.DataFrame()
```

```python
plt = sns.jointplot(x='PER_CAPITA_INCOME',y='HARDSHIP_INDEX', data=data);
```

As Per Capita Income rises, the Hardship Index decreases. The variables are negatively correlated since the points on the scatter plot are somewhat closer to a straight line in the negative direction.

### Problem 9

##### What are the top 10 schools with the highest "Average Student Attendance"?

```python
# python

highest_attendance = school_data.sort_values(by = 'AVERAGE_STUDENT_ATTENDANCE', ascending = False)
highest_attendance_table = highest_attendance[['NAME_OF_SCHOOL', 'AVERAGE_STUDENT_ATTENDANCE']].reset_index(drop = True).head(10)
highest_attendance_table
```

### Problem 11

##### Now remove the '%' sign from the above result set for Average Student Attendance column

```python
# python 

school_data['AVERAGE_STUDENT_ATTENDANCE'] = school_data['AVERAGE_STUDENT_ATTENDANCE'].str.replace('%','').astype(float)
```

```python
%%sql 

SELECT Name_of_School, REPLACE(Average_Student_Attendance, '%', '') as Average_Student_Attendance
from CHICAGO_PUBLIC_SCHOOLS_DATA 
order by Average_Student_Attendance
LIMIT 5
```

### Problem 12

##### Which Schools have Average Student Attendance lower than 70%?

```python
# python 

attend_70 = school_data[school_data['AVERAGE_STUDENT_ATTENDANCE'] < 70]
attend_70_table = attend_70[['NAME_OF_SCHOOL', 'AVERAGE_STUDENT_ATTENDANCE']].sort_values(
    by = 'AVERAGE_STUDENT_ATTENDANCE', ascending = True).reset_index(drop = True)
attend_70_table
```

```python
%%sql 

SELECT Name_of_School, Average_Student_Attendance  
from CHICAGO_PUBLIC_SCHOOLS_DATA 
where CAST (REPLACE(Average_Student_Attendance, '%', '') AS DOUBLE) < 70 
order by Average_Student_Attendance
```

### Problem 27

##### Use a sub-query to determine the Community Area Name with most number of crimes?

```python
%%sql 

SELECT community_area_name
FROM CENSUS_DATA
WHERE community_area_number = (
    SELECT community_area_number
    FROM CHICAGO_CRIME_DATA
    GROUP BY community_area_number
    ORDER BY COUNT(*) DESC
    LIMIT 1
)
```
